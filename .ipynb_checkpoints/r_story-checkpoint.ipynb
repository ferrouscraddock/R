## Step 5: Filter Results

```r
#read the data in
df <- data.frame(read.csv('/dsa/home/craddocka/fall17dsa7600_craddocka/modules/module8/exercises/311_Service_Requests_from_2014.csv'))
```

```r
library(tidyr)
library(dplyr)
library(tidyverse)
library(stringr)
library(magrittr)
library(ggplot2)
library(ggmap)
```

```r
names(df)
```

```r
#I don't need all of these variables, so I will trim them down a bit
df2 <- c('Complaint.Type','Descriptor','City','Borough','Latitude','Longitude')
df3 <- df[df2]
```

```r
#Since I'm only interested in complaints that deal with noise, I am filtering the dataset further.
NYC.noise <- df3 %>%
    filter(str_detect(Complaint.Type, 'Noise'))
head(NYC.noise)
```

```r
#I want to check for any missing values and remove if they exist.
apply(is.na(NYC.noise), 2, any)
```

```r
#Removing the NA's.
NYC.noise2 <- na.omit(NYC.noise)
```

```r
#One more check to be sure
apply(is.na(NYC.noise2), 2, any)
```

## Step 6: Visuals 

```r
#Try a quick scatter plot.
plt <- ggplot(data=NYC.noise2,aes(x=Longitude,y=Latitude)) + geom_point(aes(color=Borough)) +
    labs(title="Noise in the City",x="Longitude", y="Latitude")
plt
#Already taking on the shape of the five boroughs.
```

```r
#Gonna remove "Unspecified". Even though based on the coordinates and the plotting it is clear the reports come from Staten Island. Just want to clean things up a bit.
NYC.noise3 <- subset(NYC.noise2, Borough!="Unspecified")
```

```r
#Customize the color a little
color.scheme <- c("#f29729","#315cea","#862ce8","#d1d806","#37c2e5")
```

```r
plt <- ggplot(data=NYC.noise3,aes(x=Longitude,y=Latitude)) + 
    geom_point(aes(colour=Borough),alpha=0.2) +
    scale_color_manual(values=color.scheme) +
    labs(title="Noise in the City") + 
    theme_void() + 
    theme(legend.position = "bottom") +
    theme(plot.title = element_text(hjust = 0.5))
    
plt
```

Doesn't look too bad. It looks like Manhattan may be the noisiest borough, but it's hard to tell. I want to layer this on top of a map of the 5 boroughs to see if this will help.

```r
m <- get_map("New York City",zoom=11,maptype="toner",source="stamen")
```

```r
ggmap(m) + geom_point(aes(x=Longitude,y=Latitude,color=Borough,alpha=0.1),data=NYC.noise3) +  
    scale_color_manual(values=color.scheme) +
    labs(title="Noise in the City") +
    theme(axis.text.x = element_blank(),
    axis.text.y = element_blank(),
    axis.ticks = element_blank()) +
    theme(legend.position = "bottom") +
    theme(plot.title = element_text(hjust = 0.5)) +
    scale_alpha(guide = 'none')
```

Again, Manhattan looks to be the noisiest, but with all of the complaints crowded together into the smallest borough, is it really? I think I'll try a heat map to clear things up a little.

```r
g <- ggmap(m)
g <- g+stat_density2d(aes(x = Longitude, y = Latitude, fill=..level..), data=NYC.noise3,geom="polygon", alpha=0.2)
g + scale_fill_gradient(low = "yellow", high = "red") + 
    labs(title="Noise in the City") +
    theme(axis.text.x = element_blank(),
    axis.text.y = element_blank(),
    axis.ticks = element_blank()) +
    theme(legend.position = "bottom") +
    theme(plot.title = element_text(hjust = 0.5))
```

Much better. This clearly shows that Manhattan is the noisiest borough.

```r
Noise_Map <- get_map(location = "Manhattan,NY", zoom = 12, maptype = 'roadmap', color = 'bw')
```

```r
ggmap(Noise_Map, extent = "device") + geom_density2d(data = NYC.noise3, 
    aes(x = Longitude, y = Latitude), size = 0.3) + stat_density2d(data = NYC.noise3, 
    aes(x = Longitude, y = Latitude, fill = ..level.., alpha = ..level..), size = 0.01, 
    bins = 16, geom = "polygon") + scale_fill_gradient(low = "green", high = "red") + 
    scale_alpha(range = c(0, 0.3), guide = FALSE) +
    labs(title="Noise in the City") +
    theme(legend.position = "bottom") +
    theme(plot.title = element_text(hjust = 0.5))
```

Gonna try a map with less contrast. This looks a little cleaner and easier on the eyes. Zooming in a little shows that the noisiest part of Manhattan is just north of Washington Heights in Inwood.

```r
#And here we have it, running a table function on NYC.noise3 shows that Manhattan
#is the noisiest borough.
table(NYC.noise3$Borough)
```
